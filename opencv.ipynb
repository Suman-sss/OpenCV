```
import cv2
from google.colab.patches import cv2_imshow
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
img = cv2.imread('/content/drive/My Drive/family.jpg')
# img = cv2.resize(img,(1900,400))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=5)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),
                 (x+w,y+h),
                 color = (255,0,0),
                 thickness=2)
    cv2.putText(img,'FACE',
               (x,y+h+30),
               cv2.FONT_HERSHEY_SIMPLEX,0.6,
               (255,255,255),2)



cv2_imshow(img)
cv2.waitKey(0)
```

```
import cv2 as cv
from google.colab.patches import cv2_imshow
body_cascade = cv.CascadeClassifier(cv.data.haarcascades + "haarcascade_fullbody.xml")
img = cv.imread('/content/drive/My Drive/full.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
bodies = body_cascade.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=0)
for (x,y,w,h) in bodies:
    cv.rectangle(img,(x,y),
                 (x+w,y+h),
                 color = (255,0,0),
                 thickness=2)
    cv.putText(img,'Fullbody',
               (x,y+h+30),
               cv.FONT_HERSHEY_SIMPLEX,0.6,
               (255,255,255),2)



cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
